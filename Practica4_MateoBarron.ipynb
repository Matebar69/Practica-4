```python
import csv
from collections import defaultdict

with open('children-living-with-hiv.csv', newline='') as f:
    reader = csv.reader(f)
    header = next(reader) 
    data = list(reader)

pais_datos = defaultdict(list)
for row in data:
    pais = row[0]
    try:
        valor = float(row[3])
        pais_datos[pais].append(valor)
    except ValueError:
        continue 

filas = len(data)
columnas = len(data[0])
print(f"Total de filas: {filas}, Total de columnas {columnas}")
resultados = []
for pais, valores in pais_datos.items():
    suma = sum(valores)
    media = round(suma / len(valores), 2)
    maximo = max(valores)
    minimo = min(valores)
    resultados.append((pais, suma, media, maximo, minimo))
    
    print(f"{pais} → Suma: {suma}, Media: {media}, 'Valor Máximo: {maximo}, 'Valor mínimo: {minimo}")
    
    bolivia_anual = defaultdict(float)


for row in data:
    pais = row[0]
    if pais == "Bolivia":
        try:
            year = int(row[2])
            valor = float(row[3])
            bolivia_anual[year] += valor
        except ValueError:
            continue

print()
print("Conteo anual de niños (0-14) con VIH en Bolivia:")
for year in sorted(bolivia_anual):
    print(f"Año {year}: {int(bolivia_anual[year])} niños con VIH en Bolivia")
    



with open('resultado_hiv.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Pais', 'Suma', 'Media'])
    writer.writerows(resultados)


        
```
